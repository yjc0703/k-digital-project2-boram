# Q4 - 나의 스마트폰 과의존에 대한 인식

```python
df
```

```python
smart_use=df[['Q4']]
```

```python
smart_use
```

```python
smart_use_df11 = smart_use[['Q4']]
smart_use_df11['CNT'] = 1
```

```python
smart_use_df11
```

```python
smart_use_df11_grouped = smart_use_df11.groupby('Q4').sum()
smart_use_df11_grouped
```

```python
smart_use_df11_grouped = smart_use_df11_grouped.reset_index().drop('Q4', axis=1)
```

```python
smart_use_df11_grouped
```

```python
smart_use_df11_grouped['REASON'] = \
[
    '전혀 의존적이지 않다',
    '별로 의존적이지 않다',
    '비슷하게 의존적이다',
    '좀 더 의존적이다',
    '매우 의존적이다'
]
```

```python
smart_use_df11_grouped
```

```python
total3 = smart_use_df11_grouped['CNT'].sum()
smart_use_df11_grouped['RATE'] = ['{0:.2f}'.format(c / total3 * 100) for c in smart_use_df11_grouped['CNT']]
smart_use_df11_grouped
```

```python
mp.pyplot.pie(x=smart_use_df11_grouped['CNT'], labels=smart_use_df11_grouped['RATE'])
plt.legend(smart_use_df11_grouped['REASON'])
```

```python
# plt.pie(데이터, 라벨, autopct(전체대비비율)=소수점표시, startangle(부채꼴 시작 각도), shadow(그림자 효과 유무), explode(돌출))

plt.figure(figsize=(10,10))
plt.rc('font', size=12)

ratio = smart_use_df11_grouped['RATE'].tolist()
labels = ['전혀 의존적이지 않다','별로 의존적이지 않다','비슷하게 의존적이다','좀 더 의존적이다','매우 의존적이다']
explode = [0.03,0.03,0.03,0.03,0.03]
colors = ['#3A5363','#87BC46','#F8C11E','#8D6242','#D72428'] 
wedgeprops={'width': 0.7, 'edgecolor': 'w', 'linewidth': 2}

plt.pie(ratio,labels=labels,autopct='%.1f%%', startangle=-10, counterclock=False, explode=explode, colors=colors, wedgeprops=wedgeprops)
plt.show()
```
